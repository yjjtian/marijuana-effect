### QUESTID2 (RESPONDENT IDENTIFICATION) (codebook 3)
a unique record identifier within a particular study year. The IDs may be repeated in other study years, so analysts should always consider the study year when using the QUESTID2. It should be noted that this variable was referred to as QUESTID in prior years.
